```python
import matplotlib.pyplot as plt

from mlhep.algorithms import OptimizationAlgorithms, DEParams
```

## DIfferential Evolution

```python
algorithms = OptimizationAlgorithms()
```

```python
my_params = DEParams(
    mut = 0.8,
    crossp = 0.7,
    popsize = 20,
    its = 3000
)
```

Uni-dimensional setting: $f(x) = x^2$

```python
function = lambda x: x**2
```

```python
list(algorithms.differential_evolution(f_x=function, bounds=[(-100, 100)], params=my_params))[-1]
```

Multi-dimensional setting: $f(x) = \sum_i^n{x_i^2 / n}$ for n=32 dimensions.

```python
function2 = lambda x: sum(x**2) / 32
```

```python
algorithms = OptimizationAlgorithms()
result = list(algorithms.differential_evolution(f_x = function2, bounds=[(-100, 100)] * 32, params=my_params))
print(result[-1])
```

```python
x, f = zip(*result)
plt.plot(f)
```

```python
for d in [8, 16, 32, 64]:
    it = list(algorithms.differential_evolution(f_x=lambda x: sum(x**2)/d, bounds=[(-100, 100)] * d, params=my_params))    
    x, f = zip(*it)
    plt.plot(f, label='d={}'.format(d))
plt.legend()
plt.xlabel('Iteracin')
plt.ylabel('f(x)')
```
